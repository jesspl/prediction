```python
#5) Decidan cómo abordar los valores faltantes y cómo hacerlo. (Esto requiere sus criterios, así que expliquen su elección).

filtro=df[['Item_Weight', 'Outlet_Size', 'Item_Identifier' ]]

filtro2=filtro.sort_values(by=['Item_Identifier']) #, ascending=False

print(filtro2)


#primer criterio 

#el peso del intem parece que tiene relacion con el identrificador del item,
# por lo que se podria reemplazar el peso del producto que este vacio por los pesos del mismo identificador

#los tamaños del item no veo relacion aparente con respecto al peso

#seria mejor descartar los productos con campos vacios en los casos que no haya manera de relacionarlos

# se opto por organizar los elementos por item_identifier asi poder establecer el peso del item ya que es la relacion mas cercana


```

```python
#ejecucion del punto 5

clear=df.sort_values(by=['Item_Identifier']) #, ascending=False

clear["Item_Weight"].fillna( method ='ffill',  inplace = True)
clear[['Item_Identifier', 'Item_Weight', 'Outlet_Identifier' ]]
```

```python
df1=clear.sort_values(by=['Outlet_Identifier'])

df1["Outlet_Size"].fillna( method ='ffill',  inplace = True)

#df1[['Item_Identifier', 'Item_Weight', 'Outlet_Identifier', 'Outlet_Size' ]]
```

```python
df1
```

```python
#6) Confirmen que no hay valores faltantes después de abordarlos.

df1.isna().sum()


#Se encontraron 555 valores restantes que no hay manera de relacionar por lo que procedo a eliminarlos

df1.dropna(inplace=True)

df1
```

```python
#reverificando que no quede informacion vacia
df1.isna().sum()
```

```python
#Mapa de calor
#primero se debe encontrar la correlacin de los datos

corr = df.corr()
corr

```

```python
import seaborn as sns
sns.heatmap(corr, cmap = 'Blues', annot = True)
```

Observen que cada columna numérica se compara con la otra.  Cuando una columna se compara consigo misma, siempre es una perfecta correlación, es por eso que vemos los valores de 1 para ID con ID, edad con edad, altura con altura y año con año. 

En general, un coeficiente de correlación más grande que 0,7 se considera “fuerte”.

Entre 0,5 y 0,7 se considera “moderado”.  

Entre 0,3 y 0,5 se considera una correlación “baja”.

Sin embargo, estas reglas de oro pueden variar entre las áreas de contenido.
